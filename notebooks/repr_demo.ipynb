# `grblas` new reprs!
## Come see the new HTML repr for `grblas` objects and expressions

```python
from grblas import Vector, Matrix, binary, semiring
```

### Click on the output to show or hide the data

```python
v = Vector.from_values([0, 2, 4], [0, 1, 2])
v
```

### Give your own name to objects, and show truncated preview of large objects

```python
Vector.from_values([0, 2**59], [0, 1], name='v_huge')
```

### A Matrix and its transpose...

```python
A = Matrix.from_values([1, 3, 1, 3], [0, 1, 2, 4], [0, 1, 2, 3])
A
```

```python
# This doesn't compute the transpose (do `A.T.new()` for that).
# Instead, `A.T` is used to set the appropriate transpose descriptor when used in an expression.
A.T
```

### Masks...

```python
A.S
```

```python
~A.V
```

### Delayed expressions...

```python
A.mxv(v)
```

```python
A.mxm(A.T, semiring.min_plus)
```

```python
v.apply(binary.times, right=10)
```

```python
A.reduce_scalar()
```
